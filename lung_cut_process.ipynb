```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
import os
from scipy.signal import resample
```

```python
sound_folder_path = "E:/lung_sound/data goc"
text_folder_path = "E:/lung_sound/data text"
print(sound_folder_path)
print(text_folder_path)

both_folder_path = "E:/lung_sound/both"
crackles_folder_path = "E:/lung_sound/crackles"
wheezes_folder_path = "E:/lung_sound/wheezes"
normal_folder_path = "E:/lung_sound/normal"
```

```python
count = 0
for sound_file_name in os.listdir(sound_folder_path):
    text_file_name = sound_file_name.replace('.wav', '.txt')
    #create directory
    sound_file_path = os.path.join(sound_folder_path, sound_file_name)
    text_file_path = os.path.join(text_folder_path, text_file_name)
    #load data
    fs, sound_data = wavfile.read(sound_file_path)
    text_data = np.loadtxt(text_file_path)
    #print(sound_file_name)
    #convert 44100Hz to 4000Hz
    resampling_ratio = 4000/44100
    if fs == 44100:
        sound_data = resample(sound_data, int(len(sound_data) * resampling_ratio))
        fs = 4000
    #split text data
    text_cut_point = []
    text_cut_label = []
    for i in range(text_data.shape[0]):
        text_cut_point.append(text_data[i,0])
        text_cut_label.append((int(text_data[i,2]), int(text_data[i,3])))
    #both: (1,1), crackles: (1,0), wheezes: (0,1), normal: (0,0)
    #print(text_cut_point.shape)
    #print(text_cut_label)
    #split sound wave data
    for i in range((np.shape(text_cut_point))[0]-1):
        #Define the desired start and end times in seconds
        start_time = text_cut_point[i]  # Start time in seconds
        end_time = text_cut_point[i+1]    # End time in seconds
        #Convert start and end times to sample indices
        start_index = int(start_time * fs)
        end_index = int(end_time * fs)
        #Crop the sound signal
        cropped_signal = sound_data[start_index:end_index]
        #create name and label for each cut
        cropped_signal_name = sound_file_name.rstrip('.wav') + '_P' + str(i) + '.wav'
        cropped_signal_label = text_cut_label[i]
        print(cropped_signal_label)
        #create output directory and write out wav file
        if cropped_signal_label == (0, 0):
            out_cropped_signal_dir = os.path.join(normal_folder_path, cropped_signal_name)
            #print(out_cropped_signal_dir)
        elif cropped_signal_label == (1, 0):
            out_cropped_signal_dir = os.path.join(crackles_folder_path, cropped_signal_name)
            #print(out_cropped_signal_dir)
        elif cropped_signal_label == (0, 1):
            out_cropped_signal_dir = os.path.join(wheezes_folder_path, cropped_signal_name)
            #print(out_cropped_signal_dir)
        else:
            out_cropped_signal_dir = os.path.join(both_folder_path, cropped_signal_name)
            #print(out_cropped_signal_dir)
        wavfile.write(out_cropped_signal_dir, fs, cropped_signal)
```
