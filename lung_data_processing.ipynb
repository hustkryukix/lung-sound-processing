```python
import numpy as np
import numpy.fft as fft
import matplotlib.pyplot as plt
import os
from scipy.io import wavfile
from scipy.signal import spectrogram
import emd
```

```python
# Load the sound wave data
fs, signal = wavfile.read("E:/lung_sound/wheezes/103_2b2_Ar_mc_LittC2SE_P0.wav")  
# Replace 'sound_wave.csv' with the filename or path to your sound wave file
```

```python
imf = emd.sift.sift(signal)
print(imf.shape)
```

```python
imf_n = np.concatenate(imf,imf[0:(15000-imf.shape[0])])
```

```python
sample_rate = 1000
```

```python
IP, IF, IA = emd.spectra.frequency_transform(imf_n, sample_rate, 'hilbert')
```

```python
freq_range = (0.1, 10, 80, 'log')
f, hht = emd.spectra.hilberthuang(IF, IA, freq_range, sum_time=False)
```
